```python
stopWords.append('shall')
print(stopWords)
```

## Clean text and remove stop words function

```python
import re
```

```python
example = 'yucky-. text,'
```

```python
clean_word = re.sub(r'[^a-z ]+', '', example)
```

```python
clean_word
```

```python
for word in example_words:
    clean_word = re.sub(r'[^a-z ]+', '', word)
    if clean_word not in stopWords:
        print(clean_word)
```

```python
def clean_remove_stopwords(text):
    
    word_list = text.split()
    clean_word_list = []
    
    for word in word_list:
        lower_word = word.lower()
        clean_word = re.sub(r'[^a-z ]+', '', lower_word)
        
        if clean_word not in stopWords:
            clean_word_list.append(clean_word)
    
    return clean_word_list
```

```python
words = clean_remove_stopwords(text)
```

```python
len(words)
```

```python
words[:20]
```

```python
string_words = " ".join(words)
string_words[:200]
```

## Create word cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(background_color="grey",
               height=1000,
                width=1000,
                colormap='plasma',
                max_words=300,
                relative_scaling=0.5,
                normalize_plurals=False,
                mask=None)

word_cloud = wc.generate(string_words)
```

You can find different colormaps here: https://matplotlib.org/users/colormaps.html

```python
type(word_cloud)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## Word Cloud with Mask

```python
import numpy as np
from PIL import Image

cloud_mask = np.array(Image.open("../static/female-portrait2.png"))

wc = WordCloud(background_color="white",
                colormap='plasma',
                max_words=1000,
                mask=cloud_mask)

word_cloud = wc.generate(string_words)

word_cloud.to_file('word_cloud.png')

plt.figure(figsize=(12,12))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
